```python
def entropy(p):
    return -(p * np.log2(p) + (1-p) * np.log2((1-p)))
```

```python
#Entropy(node-caps)
entropy_node_caps = entropy(56/278)
entropy_node_caps
```

```python
#Entropy(node-caps, tumor-size)
entropy_nc_ts = ((8/278)*0.0) + ((4/278)*0.0) + ((28/278)*entropy(1/28)) + ((29/278)*entropy(4/29)) + ((48/278)*entropy(9/48)) + ((51/278)*entropy(10/51)) + ((58/278)*entropy(16/58)) + ((19/278)*entropy(7/19)) + ((22/278)*entropy(6/22)) + ((3/278)*entropy(1/3)) + ((8/278)*entropy(2/8))
```

```python
#information gain(node-caps, tumor-size)
gain_nc_ts = entropy_node_caps - entropy_nc_ts
gain_nc_ts
```
