```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.nba.com/players")

soup = BeautifulSoup(page.content, 'html')
```

```python
soup
```

```python
table = soup.find('table')
table
```

```python
table =soup.select('table')[0]
rows = table.find_all('tr')
output = []
for row in rows:
    cols = row.find_all('td')
    item = []
    for data in cols:
        item.append(data.text)
    output.append(item)


df = pd.DataFrame(output, columns = ['Player Name','Team','Number','Position','Height','Weight','Last Attended','Country'])
df = df.drop([0])
df
```
