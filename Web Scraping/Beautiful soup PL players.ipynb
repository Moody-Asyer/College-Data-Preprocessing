```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

page = requests.get("https://www.premierleague.com/players")

soup = BeautifulSoup(page.content, 'html')
```

```python
soup
```

```python
body = soup.find('tbody', {'class':'dataContainer indexSection'})
body
```

```python
body.find('a').text
```

```python
#find all disimpan dalam list sedang find tidak
player_names = []
All = soup.find('tbody', {'class':'dataContainer indexSection'})
len_data = len(All.find_all('a'))
for i in range(0,len_data):
    player_names.append(All.find_all('a')[i].text)
    print(All.find_all('a')[i].text)
```

```python
positions = []
countries = []
item = soup.find_all('td', {'class':'hide-s'})
len_data = len(item)
for i in range(0,len_data):
    if i%2 == 0:
        positions.append(item[i].text)
        print('position:', item[i].text)
    elif i%2 == 1:
        countries.append(item[i].find('span',{'class':'playerCountry'}).text)
        print('country:', item[i].find('span',{'class':'playerCountry'}).text)
        print('-----')
```

```python
data = {'player_names' : player_names,
        'positions' : positions,
        'countries' : countries}
PL = pd.DataFrame(data, columns=['player_names','positions', 'countries'])
```
