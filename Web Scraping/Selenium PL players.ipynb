```python
from selenium import webdriver
import pandas as pd
import datetime
import numpy as np
import webdriver_manager
import time
```

```python
url = 'https://www.premierleague.com/players'
```

```python
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get(url)
```

```python
time.sleep(2)  
scroll_pause_time = 1
screen_height = driver.execute_script("return window.screen.height;")
i = 1

while True:
    driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))  
    i += 1
    time.sleep(scroll_pause_time)
    scroll_height = driver.execute_script("return document.body.scrollHeight;")  
    if (screen_height) * i > scroll_height:
        break 
```

```python
def parse_table(val):
    value = driver.find_elements_by_xpath('//*[@id="mainContent"]/div[2]/div[1]/div/div/table/tbody/tr/td[{}]'.format(str(val)))
    added_value = []
    for item in value:
        member = item.text
        added_value.append(member)
        if len(added_value) == 60:
            break
    return added_value

players_name = parse_table(1)
positions = parse_table(2)
Countries = parse_table(3)
```

```python
complete_data = {"Players_Name":players_name, 
                 "Position":positions, 
                 "Country":Countries}

df = pd.DataFrame(complete_data)
df
```
